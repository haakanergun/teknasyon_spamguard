### 2.6 En Sık Kullanılan Kelimeler Analizi <a class="anchor" id="26-en-sik-kullanilan-kelimeler-analizi"></a>

Spam ve ham mesajlarda en sık kullanılan kelimeleri analiz ederek, mesajların dil yapısı hakkında fikir edineceğiz. Bu aşamada mesajları tokenize edip, stop-word'leri çıkardıktan sonra en sık geçen kelimeleri sayacağız.


```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter

# Gerekli NLTK paketlerini indiriyoruz (önceden indirilmediyse)
nltk.download('punkt_tab')
nltk.download('stopwords')

# İngilizce stop-word'ler
stop_words = set(nltk.corpus.stopwords.words('english'))

# Spam mesajlar için en sık kullanılan kelimeleri bulma
spam_words = []
for message in df[df['Label'] == 'spam']['Message']:
    tokens = word_tokenize(message.lower())
    spam_words.extend([token for token in tokens if token.isalpha() and token not in stop_words])

spam_freq = Counter(spam_words)
print("Spam Mesajlarda En Sık Kullanılan 10 Kelime:")
print(spam_freq.most_common(10))

# Ham mesajlar için en sık kullanılan kelimeleri bulma
ham_words = []
for message in df[df['Label'] == 'ham']['Message']:
    tokens = word_tokenize(message.lower())
    ham_words.extend([token for token in tokens if token.isalpha() and token not in stop_words])

ham_freq = Counter(ham_words)
print("\nHam Mesajlarda En Sık Kullanılan 10 Kelime:")
print(ham_freq.most_common(10))
```

## Adım 3: Veri Temizleme ve Metin Ön İşleme <a class="anchor" id="adim-3-veri-temizleme-ve-metin-on-isleme"></a>


### 3.1 Metin Temizleme Fonksiyonu Oluşturma <a class="anchor" id="31-metin-temizleme-fonksiyonu-olusturma"></a>

Bu bölümde, metin verilerindeki gereksiz karakterleri, noktalama işaretlerini ve stop-word’leri (hem İngilizce hem Türkçe) temizleyecek bir fonksiyon oluşturacağız.  
Fonksiyonumuz:
- Metni küçük harfe çevirir,
- Türkçe karakterler de dahil olmak üzere yalnızca alfanümerik karakterler ve boşlukları korur,
- Fazla boşlukları tek boşluğa indirger,
- Token’lara ayırır ve stop-word’leri temizler.


```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Türkçe stop-word'lerden oluşan kısmi bir liste (isteğe bağlı olarak geliştirilebilir)
turkish_stopwords = {
    "acaba", "ama", "aslında", "az", "bazı", "belki", "biri", "birkaç", "birçok",
    "bu", "çok", "çünkü", "da", "de", "daha", "eğer", "en", "gibi", "hem", "hep",
    "hepsi", "için", "ile", "ise", "kim", "mı", "mu", "mü", "nasıl", "ne", "neden",
    "niçin", "o", "sanki", "şey", "şu", "tüm", "ve", "veya", "ya", "yani"
}

# İngilizce stop-word listesi (NLTK'dan)
english_stopwords = set(stopwords.words('english'))

# İki dildeki stop-word'leri birleştirelim
all_stopwords = english_stopwords.union(turkish_stopwords)

def clean_text(text):
    """
    Verilen metni:
    - Küçük harfe çevirir,
    - Türkçe karakterleri de koruyarak alfanümerik karakterler dışındaki karakterleri temizler,
    - Fazla boşlukları indirger,
    - Token’ları oluşturup stop-word’leri temizler.
    """
    # Küçük harfe çevirme
    text = text.lower()
    
    # Türkçe karakterleri de içeren alfabeye uygun regex ifadesi:
    # a-z, 0-9, ç, ğ, ı, ö, ş, ü ve boşluklar dışındaki karakterleri temizler
    text = re.sub(r'[^a-z0-9çğıöşü\s]', ' ', text)
    
    # Fazla boşlukları tek boşluk haline getirme
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Tokenize etme
    tokens = text.split()
    
    # Stop-word'leri kaldırma
    tokens = [word for word in tokens if word not in all_stopwords]
    
    # Temizlenmiş metni tekrar birleştirip döndürme
    return " ".join(tokens)

# Örnek bir mesaj üzerinde fonksiyonu test edelim
example_text = "Selamat pagi pak, untuk data table yg dibutuhkan sudah saya email. Apakah email saya msk Pak?"
print("Orijinal Mesaj:")
print(example_text)
print("\nTemizlenmiş Mesaj:")
print(clean_text(example_text))
```

### 3.2 Temizlenmiş Mesaj Sütunu Oluşturma <a class="anchor" id="32-temizlenmis-mesaj-sutunu-olusturma"></a>

Oluşturduğumuz `clean_text` fonksiyonunu, tüm veri setimizdeki "Message" sütununa uygulayarak yeni bir sütun (`Clean_Message`) oluşturacağız.  
Bu sütun, modelleme aşamasında kullanılacak metin verisinin ön işlenmiş halini içerecek.


```python
# "Message" sütunundaki tüm metinleri temizleyip, yeni sütuna atama
df['Clean_Message'] = df['Message'].apply(clean_text)

# İlk 5 satırda orijinal ve temizlenmiş mesajları karşılaştıralım
df[['Message', 'Clean_Message']].head()
```

## Adım 4: Özellik Çıkarımı <a class="anchor" id="adim-4-ozellik-cikarimi"></a>


### 4.1 TF-IDF Vektörleştirme <a class="anchor" id="41-tf-idf-vektorlestirme"></a>

Bu adımda, `Clean_Message` sütunundaki ön işlenmiş metin verisini TF-IDF vektörleştirme yöntemiyle sayısal özelliklere dönüştüreceğiz.  
TF-IDF, metin içindeki kelimelerin önemini belirleyerek, her mesajı vektör şeklinde temsil etmemize olanak tanır.


```python
import pickle

# Örneğin, vectorizer'ınız TF-IDF vektörleştiricisi olarak tanımlı olsun:
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(df['Clean_Message'])

# Kaydetme işlemi:
with open("tfidf_vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)
```

### 4.2 Etiketlerin Kodlanması <a class="anchor" id="42-etiketlerin-kodlanmasi"></a>

Modelin anlayabilmesi için etiket sütununu sayısal değerlere dönüştürüyoruz.  
Burada "ham" mesajlar 0, "spam" mesajlar ise 1 olarak kodlanacaktır.


```python
# Etiket sütununu sayısallaştırma
y = df['Label'].map({'ham': 0, 'spam': 1})
print("Etiketlerin Dağılımı:")
print(y.value_counts())
```

## Model Sonuçlarına Genel Yorumlar

Aşağıdaki model performans tablosunda, çeşitli sınıflandırma algoritmalarının SMS spam tespiti üzerinde elde ettikleri sonuçlar yer alıyor:

- **SVM (Linear Kernel) ve Ridge Classifier:**
  - **Performans:** Her iki model de %98 civarında yüksek doğruluk (Accuracy ~0.9839 ve 0.9835) ve AUC değerleri (0.9972 ve 0.9973) sunuyor.
  - **Recall ve Precision:** SVM %96.63 recall ve %98.57 precision, Ridge ise %96.11 recall ve %98.97 precision değerleriyle neredeyse mükemmel sonuçlar veriyor. Böylece her iki model de yanlış negatif ve yanlış pozitif oranlarını minimumda tutuyor.
  - **İnce Not:** SVM, recall yönünden biraz daha öne çıkarken, Ridge modelinde precision değeri biraz daha yüksek. Eğitim süreleri de oldukça kısa (1.856 ve 2.279 saniye), bu da gerçek zamanlı uygulamalarda tercih sebebi olabilir.

- **Extra Trees ve Random Forest:**
  - **Performans:** Her iki model de %98 civarında doğruluk ve yüksek AUC değerleri (0.9962 civarı) elde etmiş durumda. Extra Trees modelinde recall %96.26, Random Forest’ta ise %95.78; precision değerleri de sırasıyla %97.83 ve %98.11.
  - **F1, Kappa ve MCC:** F1 skoru, Kappa ve MCC değerleri de Extra Trees (%97.04, %95.54, %95.55) ve Random Forest (%96.93, %95.39, %95.41) için yüksek çıkmış.
  - **Hesaplama Süresi:** Extra Trees modeli, biraz daha uzun (7.350 saniye) eğitim süresine sahipken, Random Forest 4.531 saniyede sonuç vermiş. Bu modeller, yüksek performans sunmalarına rağmen hesaplama süresi açısından SVM ve Ridge’e göre daha maliyetli olabilir.

- **LightGBM, Logistic Regression ve XGBoost:**
  - **Performans:** Bu modeller de %97 civarında doğruluk ve AUC değerleri (LightGBM için 0.9926, Logistic Regression için 0.9960, XGBoost için 0.9914) ile iyi sonuçlar veriyor. Ancak, özellikle recall oranları (LightGBM %94.27, LR %92.56, XGBoost %92.49) biraz daha düşük.
  - **Genel Değerlendirme:** Bu modeller, yorumlanabilirlik ve eğitim süresi açısından (örneğin, LR 2.681 saniye, XGBoost 5.188 saniye) tercih edilebilir. Logistic Regression, özellikle modelin anlaşılabilirliğinin önemli olduğu durumlarda avantaj sağlayabilir.

- **Naive Bayes:**
  - **Performans:** %96.64 doğruluk ve %96.52 AUC gibi iyi sonuçlar elde etmiş olsa da, precision ve F1 skorları biraz daha düşük (sırasıyla %94.09 ve %95.08).
  - **Yorum:** Metin sınıflandırması için geleneksel bir yöntem olan Naive Bayes, özellikle basitlik ve hızlı hesaplama süresi (1.610 saniye) ile öne çıksa da, diğer gelişmiş modellerin biraz gerisinde kalıyor.

- **Decision Tree ve AdaBoost:**
  - **Performans:** Decision Tree ve AdaBoost modelleri, sırasıyla %95.93 ve %95.54 doğruluk elde etmiş; ancak diğer modellerle kıyaslandığında recall ve F1 skorları daha düşük. AdaBoost modelinde %90.42 recall, Decision Tree’de ise %93.19 recall gözlemleniyor.
  - **Yorum:** Bu modeller, sınırlı sayıda özellik ile çalışırken daha basit yapıya sahip oldukları için, veri setindeki karmaşık örüntüleri yakalamada zorlanabiliyor.

- **Gradient Boosting Classifier:**
  - **Performans:** %93.71 doğruluk ve %98.22 AUC elde etse de, recall değeri (%82.80) oldukça düşük. Bu durum, spam sınıfını kaçırma olasılığının daha yüksek olduğunu gösteriyor.
  - **Yorum:** Yüksek precision (%98.35) değeri, doğru pozitif oranının yüksek olduğunu gösterse de, düşük recall nedeniyle spam tespitinde eksik kalabilir.

- **Linear Discriminant Analysis (LDA) ve Quadratic Discriminant Analysis (QDA):**
  - **Performans:** LDA %87.09 doğruluk, QDA ise sadece %72.01 doğruluk sunuyor. Özellikle QDA’nın AUC değeri (%9.59) çok düşük; bu modeller, spam tespiti için uygun değil.
  - **Yorum:** Bu modeller, veri setindeki dağılım varsayımlarını yeterince karşılamadığı için düşük performans gösteriyor.

- **Dummy Classifier:**
  - **Performans:** %66.21 doğruluk ve %50 AUC ile, rastgele tahmin yapan bu model referans (baseline) olarak kullanılıyor.
  - **Yorum:** Bu model, diğer modellerin performansını kıyaslamak için temel alınmıştır. Hiçbir akıllı model bu sonuçlara yaklaşamaz.

### Genel Değerlendirme:

- **Öne Çıkan Modeller:**  
  SVM, Ridge Classifier, Extra Trees ve Random Forest; yüksek doğruluk, AUC, recall, precision, F1, Kappa ve MCC değerleri ile spam tespiti konusunda mükemmel sonuçlar sunuyor. Ayrıca SVM ve Ridge, kısa eğitim süreleri ile üretim ortamında tercih edilebilir.

- **Kullanım Senaryoları:**

  - **Ensemble Yöntemleri:** Extra Trees ve Random Forest, yüksek doğruluk ve güvenilirlik sağlar; ancak hesaplama süresi daha yüksek olabilir.
  - **Basitlik ve Yorumlanabilirlik:** Logistic Regression, performans açısından rakipleri kadar yüksek olmasa da, yorumlanabilirliği sayesinde tercih edilebilir.
  
- **Dikkat Edilmesi Gerekenler:**  
  - **Model Seçimi:** Verinin doğası, üretim ortamındaki hesaplama gücü ve uygulamanın gereksinimlerine göre model seçimi yapılmalıdır.

  - **Gerçek Zamanlı Uygulamalar:** Eğitim süresi (TT) ve model boyutu da göz önünde bulundurularak, uygulamanın ihtiyaçlarına en uygun model tercih edilmelidir.

Bu sonuçlar, SMS spam tespiti için uygulanabilecek en uygun modellerin belirlenmesinde yol gösterici olacaktır. - **Recall (Duyarlılık):**  
  - Ortalama Recall: 0.9748

Fold’lar arasında küçük bir varyasyon bulunması, bazı fold’larda spam dağılımının biraz farklı olabileceğini düşündürse de genel performans oldukça istikrarlı.

- **Precision (Kesinlik):**  
  - Ortalama Precision: 0.9590  
  - Standart sapma: 0.0073  
  - Yorum: Modelin, pozitif sınıf olarak tahmin ettiği mesajların %95.90’ı doğru olarak sınıflandırılmıştır. Precision ve recall arasındaki uyum, dengeli bir performans sergilediğini göstermektedir.

- **F1-Score:**  
  - Ortalama F1: 0.9668  
  - Standart sapma: 0.0048  
  - Yorum: F1 skoru, hem precision hem de recall’ün dengeli olduğunu vurgular. %96.68’lik F1 skoru, modelin genel performansının yüksek olduğunu ortaya koymaktadır.

- **Kappa ve MCC (Cohen’s Kappa ve Matthews Correlation Coefficient):**  
  - Ortalama Kappa: 0.9496, Ortalama MCC: 0.9498  
  - Standart sapma: Her iki metrik için de yaklaşık 0.0072  
  - Yorum: Hem Kappa hem MCC değerleri 0.94 ve üzeri çıktığı için, modelin sınıflar arasındaki uyum ve korelasyonunun çok yüksek olduğunu söyleyebiliriz. Bu, modelin tahminlerinin rastgelelikten çok uzak, güvenilir olduğunu göstermektedir.

- **1320**: Gerçekte “ham” olan 1320 mesaj doğru şekilde “ham” olarak sınıflandırılmış (**True Negative**).  
- **4**: Gerçekte “ham” olan 4 mesaj yanlışlıkla “spam” olarak sınıflandırılmış (**False Positive**).
